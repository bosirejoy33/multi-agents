# NaiWatches Multi-Agent Demo
This notebook demonstrates the collaborative workflow of the NaiWatches Agent Grid.

```python
import google.generativeai as genai
# genai.configure(api_key='YOUR_KEY_HERE')
model = genai.GenerativeModel('gemini-1.5-flash')
```

## 1. Agent Functions
Defining the specialized workers.

```python
def run_workflow(query):
    print(f'User Query: {query}')
    
    # 1. Profiler Agent
    profile = model.generate_content(f'Analyze user profile: {query}').text
    print(f'\n[Vibe Analyzer]: Profile Created.')
    
    # 2. Researcher Agent
    research = model.generate_content(f'Research movies for: {profile}').text
    print(f'[Web Scout]: Data Gathered.')
    
    # 3. Critic Agent
    critique = model.generate_content(f'Critique these recs: {research} for {profile}').text
    print(f'[Hype Checker]: Evaluation Complete.')
    
    return research, critique
```

```python
results, feedback = run_workflow('I want some high-octane 90s action movies like Heat.')
print('\n--- FINAL RESULTS ---')
print(results)
```
