```python
def run_workflow(query):
    print(f'User Query: {query}')
    
    # 1. Profiler Agent
    profile = model.generate_content(f'Analyze user profile: {query}').text
    print(f'\n[Vibe Analyzer]: Profile Created.')
    
    # 2. Researcher Agent
    research = model.generate_content(f'Research movies for: {profile}').text
    print(f'[Web Scout]: Data Gathered.')
    
    # 3. Critic Agent
    critique = model.generate_content(f'Critique these recs: {research} for {profile}').text
    print(f'[Hype Checker]: Evaluation Complete.')
    
    return research, critique
```
